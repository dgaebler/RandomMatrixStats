# Introduction: Numerical Range Geometry
In this "second generation" pass at the case of a $2 \times 2$ matrix of real iid $U[0,1]$ entries, we try to efficiently cut to formulas for the pdf's of the major and minor axes and co-eccentricity.

Per the Elliptical Range Theorem for matrix entries, given $\begin{bmatrix} a & b \\ c & d \end{bmatrix}$ the geometric properties of the numerical range can be described as follows:
- The major axis is $\sqrt{(b+c)^2+(a-d)^2}$
- The minor axis is $|b-c|$
- The focal length is $|\sqrt{(a-d)^2+4bc}|/2$.  Note that absolute values are needed since the square root could be complex.
- The eccentricity is $\dfrac{|\sqrt{(a-d)^2+4bc}|}{2\sqrt{(b+c)^2+(a-d)^2}}$
- The co-eccentricity is $\dfrac{|b-c|}{\sqrt{(b+c)^2+(a-d)^2}}$

# Background: Transforming pdf's

In order to compute pdf's of the above expressions, we will need to address
- the sum of independent random variables with known pdf's
- the negative of a random variable with known pdf 
- the difference of independent random variables with known pdf's
- the absolute value of a random variable with known pdf
- the product of independent random variables with known pdf's
- the quotient of independent random variables with known pdf's
- an invertible function of a random variable with known pdf

### Sum of Independent RV's
This is the most famous: The pdf of a sum is the **convolution** of the pdf's, assuming independence.  Hence, given $f_X(x)$ and $f_Y(y)$, with $Z = X+Y$ one has
$$
f_Z(z) = f_X \ast f_Y(z) = \int_{-\infty}^\infty f_X(z-y) f_Y(y) \, dy =\int_{-\infty}^\infty f_Y(z-x) f_X(x) \, dx.
$$

### Negative of RV
Suppose $f_X(x)$ is known and $Y = -X$.  Straightforwardly, $f_Y(y) = f_X(-y)$.

### Difference of Independent RV's
A difference is a sum composed with a negative, i.e. $Y-X = Y + (-X)$.  By the previous two, then, if $Z = Y-X$ then
$$
f_Z(z) = \int_{-\infty}^\infty f_X(y-z) f_Y(y) \, dy = \int_{-\infty}^\infty f_Y(z-x) f_X(-x) \, dx.
$$

### Absolute Value of a RV
Suppose $Y = |X|$.  Then $Y$ takes on a given nonnegative value precisely when $X$ takes on either that value or its negative, so that
$$
f_Y(y) = \Big( f_X(y) + f_X(-y) \Big) \boldsymbol{1}_{[0,\infty)}
$$
where $\boldsymbol{1}_{E}$ denotes the **indicator function**
$$
\boldsymbol{1}_{E}(t) = \begin{cases} 1 & t \in E \\ 0 & t \notin E. \end{cases}
$$

### Product of Independent RV's
According to the Glen/Leemis/Drew paper, citing Rohatgi's book, the density of $V = XY$ is
$$
f_V(v) = \int_{-\infty}^\infty f_{X,Y} \left(x, \dfrac{v}{x} \right) \dfrac{1}{|x|} \, dx.
$$
In the case where $X, Y$ are independent this simplifies to
$$
f_V(v) = \int_{-\infty}^\infty f_{X}(x) f_Y \left(\dfrac{v}{x} \right) \dfrac{1}{|x|} \, dx.
$$
Other methods (e.g. Mellin transform) are common for handling products of RV's, due to this integral sometimes being infeasible, but in theory this formula does the trick.

Note that, to derive this formula, one can follow either of two approaches: Find the cdf by an appropriate double integral, then differentiate the result; or, use $\ln V = \ln X + \ln Y$ to find the density of $\ln V$, then exponentiate to $V$ itself (using the function-of-RV formula below).  Either way, it's necessary to consider cases based on the signs of $X$ and $Y$.

### Quotients of Independent RV's
The Curtiss paper says that for $Z = X/Y$, the density is
$$
f_Z(z) = \int_{-\infty}^\infty |y| f_X(zy) f_Y(y) \, dy.
$$

### Function of a RV
Assuming $g$ is invertible, with $Y = g(X)$ one has
$$
f_Y(y) = f_X(h(y)) |h'(y)| 
$$
where $h = g^{-1}$.

# Major Axis

As is standard knowledge (INSERT REFERENCE), the sum and difference of uniform random variables are triangularly distributed.  In this case
$$
f_{a+d}(t) = (1-|1-t|) \boldsymbol{1}_{[0,2]}(t)
$$
$$
f_{b-c}(t) = (1-|t|) \boldsymbol{1}_{[-1,1]}(t)
$$

It follows that
$$
f_{|b-c|}(t) = 2 f_{b-c}(t) \boldsymbol{1}_{[0,\infty)}(t) = 2 (1-t) \boldsymbol{1}_{0,1}(t).
$$

For squares, if $X = a+d$ and $Y = X^2$, then $X = \sqrt{Y}$, so $F_Y(y) = \frac{1}{2 \sqrt{y}} f_X(\sqrt{y})$. So
$$
f_{(a+d)^2}(t) = \frac{1}{2 \sqrt{t}} (1-|1-\sqrt{t}|) \boldsymbol{1}_{[0,2]}(\sqrt{t}) = \frac{1}{2 \sqrt{t}} (1-|1-\sqrt{t}|)  \boldsymbol{1}_{[0,4]}(t).
$$
Similarly, we write $(b-c)^2$ as $|b-c|^2$ because the squaring function is invertible on the positive reals; then
$$
f_{(b-c)^2}(t) = \frac{1}{2 \sqrt{t}} 2(1-\sqrt{t}) \boldsymbol{1}_{[0,1]}(\sqrt{t})
= (t^{-1/2}-1) \boldsymbol{1}_{[0,1]}(t).
$$

By convolution, then,
$$
f_{(a+d)^2 + (b-c)^2}(t) = \int_{-\infty}^\infty f_{(a+d)^2}(t-x) f_{(b-c)^2}(x) \, dx
= \int_{-\infty}^\infty \dfrac{1-|1-\sqrt{t-x}|}{2 \sqrt{t-x}} (x^{-1/2}-1) \boldsymbol{1}_{[0,4]}(t-x) \boldsymbol{1}_{[0,1]}(x) \, dx = \int_0^1 \dfrac{1-|1-\sqrt{t-x}|}{2 \sqrt{t-x}} (x^{-1/2}-1) \boldsymbol{1}_{[0,4]}(t-x)  \, dx.
$$

```python
def integrand_1(t): #major-axis density function, first piece, i.e. assume 0 <= t <= 1
    return lambda x: integrand(x,t)
def f_ma2(t): #major axis squared...supported on [0,5]
    return numerical_integral(integrand_1(t),max(0,t-4),min(1,t), eps_abs=0.001)[0] #numerical_integral returns a pair, where 2nd is tolerance
#plot(f_ma,0,5, plot_points = 200)  takes a couple minutes...uncomment to see
```

# Co-Eccentricity
To avoid complex square roots---and because, later on, we want to introduce it as a measure of non-normality---we explore co-eccentricity rather than eccentricity.  We have (at least numerically) the densities of both minor and major axes.  The formula for quotients then tells us
$$
f_{co-e}(t) = \int_{-\infty}^\infty |y| f_{min}(ty) f_{maj}(y) \, dy.
$$
We know that this is supported on $t \in [0,1]$, and also know from above that $f_{maj}$ is supported on $[0,\sqrt{5}]$.  This reduces the above to
$$
f_{co-e}(t) = 2\int_0^{\min(\sqrt{5}, 1/t)} (1-ty) f_{maj}(y) \, dy.
$$
At this point we turn to numerical evaluation.

```python
def integrand_coe(t):
    return lambda y: (1-t*y)*f_ma(y)
def f_coe(t):
    if (t<0 or t>1):
        return 0
    upper_bd = min(sqrt(5),1/t)
    return numerical_integral(integrand_coe(t),0,upper_bd, eps_abs=0.001)[0]
#f_coe(0.4) This times out.  Curse of dimensionality?
```

OK, it looks like the runtime is becoming prohibitive, even with a generous error tolerance in the numerical integration.  Time to take a step back.  Since $f_{maj}$ is itself defined as an integral, the above expression implies that $f_{co-e}$ is a double integral.  Explicitly, it turns out that
$$
f_{co-e}(t) = \int_0^{\min(\sqrt{5}, 1/t)} \int_{\max(0, \sqrt{y}-4)}^{\min(1, \sqrt{y})} \dfrac{(1-ty) (x^{-1/2}-1)\left(1-\Big|1-\sqrt{\sqrt{y}-x} \Big| \right)}{2 \sqrt{y}\sqrt{\sqrt{y}-x}} \, dx \, dy
$$

```python
from scipy import integrate as sint
```

```python
var('x y t')
def coe_inside(x,y,third):
    return (1-third*y)*(x^(-1/2)-1)*(1-abs(1-sqrt(sqrt(y)-x)))/(2*sqrt(y)*sqrt(sqrt(y)-x))
def lower_bd(y):
    return max(0, sqrt(y)-4)
def upper_bd(y):
    return min(1, sqrt(y))
def f_coe_again(t):
    return sint.dblquad(coe_inside,  0, min(sqrt(5),1/t), lower_bd, upper_bd, args=[t])[0]
#f_coe_again(0.2)
plot(f_coe_again,plot_points=15)
```
